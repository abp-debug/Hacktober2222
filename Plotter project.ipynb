```python
def createdata(data):
    if(data==1):
        x=np.random.rand(100,5)
        df1=pd.DataFrame(x,columns=['A','B','C','D','E'])
    elif(data==2):
        x=[0,0,0,0,0]
        r1=[0,0,0,0,0]
        r2=[0,0,0,0,0]
        r3=[0,0,0,0,0]
        r4=[0,0,0,0,0]
        print("Enter the values you for columns.")
        i=0
        for i in range(5):
            x[i]=input()
        print("Enter the values you for 1st row.")
        i=0
        for i in range(5):
            r1[i]=int(input())
        print("Enter the values you for 2nd row.")
        i=0
        for i in range(5):
            r2[i]=int(input())
        print("Enter the values you for 3rd row.")
        i=0
        for i in range(5):
            r3[i]=int(input())
        print("Enter the values you for 4th row.")
        i=0
        for i in range(5):
            r4[i]=int(input())
        df1=pd.DataFrame([r1,r2,r3,r4],columns=x)
    elif(data==3):
        file=input("Enter the file name")
        x=pd.read_csv(file)
        df1=pd.DataFrame(x)
    else:
        print("Data creation failed .Try again with right data.")
    
    return df1
    
```

```python
def main(cat):
    if(cat==1):
        print("Select the plot the you want to get plotted for your data")
        print("1.Line Plot\n2.Scatter Plot\n3.Bar Plot\n4.Histogram Plot\n5.Box Plot\n6.Surface Plot")
        plot=int(input())
        output=plotter(plot)
    elif(cat==2):
        print("Select the plot the you want to get plotted for your data")
        print("1.Line Plot\n2.Scatter Plot\n3.Bar Plot\n4.Histogram Plot\n5.Box Plot\n6.Surface Plot\n7.Bubble Plot")
        plot=int(input())
        output=plotter2(plot)
    else:
        print("Correctly choose the option")
    return output
```

```python
def plotter(plot):
    if(plot==1):
        finalplot=df1.iplot(kind='scatter')
    elif(plot==2):
        finalplot=df1.iplot(kind='scatter',mode= 'markers',symbol='x')
    elif(plot==3):
        finalplot=df1.iplot(kind='bar')
    elif(plot==4):
        finalplot=df1.iplot(kind='hist')
    elif(plot==5):
        finalplot=df1.iplot(kind='box')
    elif(plot==6):
        finalplot=df1.iplot(kind='surface')
    else:
        print("You enterd a wrong choice. Try again.\nEnter a right choice between 1 to 6")
    
    
    return finalplot    
```

```python
def plotter2(plot):
    col=int(input("Enter the no of column 1,2 or 3"))
    if(col==1):
        colm=input("Enter the column from the DataFrame head.")
        if(plot==1):
            finalplot=df1[colm].iplot(kind='scatter')
        elif(plot==2):
            finalplot=df1[colm].iplot(kind='scatter',mode= 'markers',symbol='x')
        elif(plot==3):
            finalplot=df1[colm].iplot(kind='bar')
        elif(plot==4):
            finalplot=df1[colm].iplot(kind='hist')
        elif(plot==5):
            finalplot=df1[colm].iplot(kind='box')
        elif(plot==6 or plot==7):
            print("Suface and the Bubble plot need more than two columns.")
        else:
            print("You enterd a wrong choice. Try again.\nEnter a right choice between 1 to 6")
    elif(col==2):
        x=input("Enter the 1st column from the DataFrame head.")
        y=input("Enter the 2nd column from the DataFrame head.")
        if(plot==1):
            finalplot=df1[[x,y]].iplot(kind='scatter')
        elif(plot==2):
            finalplot=df1[[x,y]].iplot(kind='scatter',mode= 'markers',symbol='x')
        elif(plot==3):
            finalplot=df1[[x,y]].iplot(kind='bar')
        elif(plot==4):
            finalplot=df1[[x,y]].iplot(kind='hist')
        elif(plot==5):
            finalplot=df1[[x,y]].iplot(kind='box')
        elif(plot==6):
            finalplot=df1.iplot(kind='surface',x=x,y=y)
        elif(plot==7):
            print("Enter  the size of the bubble sort as one of the column from the DataFrame head")
            size=input()
            finalplot=df1.iplot(kind='bubble',x=x,y=y,size=size)
        else:
            print("You enterd a wrong choice. Try again.\nEnter a right choice between 1 to 6")
    elif(col==3):
        x=input("Enter the 1st column from the DataFrame head.")
        y=input("Enter the 2nd column from the DataFrame head.")
        z=input("Enter the 3rd column from the DataFrame head.") 
        if(plot==1):
            finalplot=df1[[x,y,z]].iplot(kind='scatter')
        elif(plot==2):
            finalplot=df1[[x,y,z]].iplot(kind='scatter',mode= 'markers',symbol='x')
        elif(plot==3):
            finalplot=df1[[x,y,z]].iplot(kind='bar')
        elif(plot==4):
            finalplot=df1[[x,y,z]].iplot(kind='hist')
        elif(plot==5):
            finalplot=df1[[x,y,z]].iplot(kind='box')
        elif(plot==6):
            finalplot=df1.iplot(kind='surface',x=x,y=y,z=z)
        elif(plot==7):
            print("Enter  the size of the bubble sort as one of the column from the DataFrame head")
            size=input()
            finalplot=df1.iplot(kind='bubble',x=x,y=y,z=z,size=size)
        else:
            print("You enterd a wrong choice. Try again.\nEnter a right choice between 1 to 6")
    else:
        print("You entered the wrong choice. Please enter between 1 or 2 or 3")
        
```
